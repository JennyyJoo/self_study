## sklearn 라이브러리를 사용한 feature scaling

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
x=pd.read_csv("C:/Users/HYJoo/NBA_player_of_the_week.csv")
x
```

```python
x.describe()
```

```python
height_weight_age_df=x[['Height CM','Weight KG','Age']]
```

```python
height_weight_age_df
```

### minmax normalization

```python
scaler=preprocessing.MinMaxScaler() 
# feature scaling 위해 skleanr preprocessing 에서 minmaxscaler가져옴
```

```python
normalization_data=scaler.fit_transform(height_weight_age_df)
normalization_data
```

```python
normalized_df=pd.DataFrame(normalization_data, columns=['Height CM','Weight KG','Age'])
```

```python
normalized_df.describe()
```
